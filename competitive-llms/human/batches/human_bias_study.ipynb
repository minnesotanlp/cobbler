```python
def convert_amt_to_df(batch_path):

    d = pd.read_csv(batch_path)
    d = d[['AssignmentId', 'WorkerId', 'Answer.taskAnswers', 'SubmitTime', 'WorkTimeInSeconds']]
    d['WorkTime_min'] = d['WorkTimeInSeconds']/60
    d = d[['AssignmentId', 'WorkerId', 'Answer.taskAnswers', 'SubmitTime','WorkTime_min']]

    return d
```

```python
def check_invalid_data(order_path):

    all_data = convert_amt_to_df(order_path)
    raters_data = all_data['Answer.taskAnswers'].values
    invalid_worker = []
    invalid_worker_id = []

    for i, rater in enumerate(raters_data):
        rater_data = json.loads(rater)[0]
        if (len(rater_data) != 0):
            sorted_data = sorted(rater_data.items(), key=lambda x: int(x[0].split("_")[1]))
            for d, ratings in sorted_data:
                if ratings == {'A': False, 'B': False}:
                    print(f"Didn't respond to questions: {all_data['WorkerId'][i]}")
                    print(f"d: {d}, ratings: {ratings}")
                    invalid_worker.append(all_data['WorkerId'][i])
                    invalid_worker_id.append(i)
        else:
            invalid_worker.append(all_data['WorkerId'][i])
            invalid_worker_id.append(i)
            print(f"Nothing submitted: {all_data['WorkerId'][i]}")
    
    return list(set(invalid_worker)), list(set(invalid_worker_id))
```

```python
def return_valid_data(amt_path):
    amt_data = convert_amt_to_df(amt_path)['Answer.taskAnswers'].values
    amt_worker_info = convert_amt_to_df(amt_path)['WorkerId'].values
    _, invalid_workers_idx = check_invalid_data(amt_path)
    valid_raters_data = np.delete(amt_data, invalid_workers_idx) 
    valid_raters_id = np.delete(amt_worker_info, invalid_workers_idx)

    return valid_raters_data, valid_raters_id
```

```python
def show_valid_data(amt_path):

    valid_raters_data, valid_raters_id = return_valid_data(amt_path)

    for i, rater in enumerate(valid_raters_data):
        print(f"Rater idx: {i}, rater info: {valid_raters_id[i]} ")
        rater_data = json.loads(rater)[0]
        sorted_data = sorted(rater_data.items(), key=lambda x: int(x[0].split("_")[1]))
        rater_organized_data = {}
        for task, ratings in sorted_data:
            if ratings['A']:
                rater_organized_data[task] = 'A'
            elif ratings['B']:
                rater_organized_data[task] = 'B'

        print(rater_organized_data)
```

```python
def quality_check_with_me(amt_path, me_path):

    valid_raters_data, valid_raters_id = return_valid_data(amt_path)
    with open(me_path, 'r') as me:
        me_data = json.load(me)

    for i, rater in enumerate(valid_raters_data):
        print(f"Rater idx: {i}, rater info: {valid_raters_id[i]} ")
        rater_data = json.loads(rater)[0]
        sorted_data = sorted(rater_data.items(), key=lambda x: int(x[0].split("_")[1]))
        rater_organized_data = {}
        for task, ratings in sorted_data:
            if ratings['A']:
                rater_organized_data[task] = 'A'
            elif ratings['B']:
                rater_organized_data[task] = 'B'

        
        matching_tasks = sum(1 for key in me_data if me_data[key] == rater_organized_data[key])
        perc_matching = (matching_tasks)/(len(me_data))
        if perc_matching >= 0.5: 
            print(f"Rater idx {i} passed 50% similarity, showing {round(perc_matching*100, 3)}%.")
        else:
            print(f"Rater idx {i} failed 50% similarity, showing {perc_matching*100}%.")
```

```python
def compute_fleiss_kappa(amt_path):

    valid_raters_data, valid_raters_id = return_valid_data(amt_path)

    all_flattened = []

    for i, rater in enumerate(valid_raters_data):
        # print(f"Rater idx: {i}, rater info: {valid_raters_id[i]} ")
        rater_data = json.loads(rater)[0]
        sorted_data = sorted(rater_data.items(), key=lambda x: int(x[0].split("_")[1]))
        rater_organized_data = {}
        for task, ratings in sorted_data:
            if ratings['A']:
                rater_organized_data[task] = 'A'
            elif ratings['B']:
                rater_organized_data[task] = 'B'

        # print(rater_organized_data)

        labels_with_true = list(rater_organized_data.values())
        # print(labels_with_true)
        all_flattened.append(labels_with_true)
    stacked_raters = np.column_stack((all_flattened[0], all_flattened[1], all_flattened[2]))
    # print(aggregate_raters(stacked_raters))
    kappa = fleiss_kappa(aggregate_raters(stacked_raters)[0])
    return kappa


def compute_kappa_bias(bias_type):
    all_kappa = []
    for i in range(1, 26):
        path_var_name = f'{bias_type}_path_{i}'  # Format the variable name
        path_value = eval(path_var_name)   # Get the value of the variable
        kappa = compute_fleiss_kappa(path_value)
        print(f"{i}: {kappa}")
        all_kappa.append(kappa)

    return np.mean(all_kappa)

```

```python
def aggregate_annotators_across_order(batch_i):

    all_instructions_lst = []
    agg_bias = {'first-order': [], 'last_order':[]}

    # For each instruction, 
    # for batch_i in range(1, 26): 
    path_var_name = f'order_path_{batch_i}'  # Format the variable name
    path_value = eval(path_var_name)   # Get the value of the variable
    valid_raters_data, valid_raters_id = return_valid_data(path_value)

    with open('human_bias/order/real/random_sampled_examples_order_{}.json'.format(batch_i), 'r') as j:
        batch_json = json.load(j)

    instruction_dict = {}
    # For each rater (among 3),
    for i, rater in enumerate(valid_raters_data):
        rater_data = json.loads(rater)[0]
        sorted_data = sorted(rater_data.items(), key=lambda x: int(x[0].split("_")[1]))

        rater_task_answer = {}
        # Create a dictionary to store pairwise counts for each model
        rater_organized_data = dict.fromkeys(model_names, 0)

        # For each of 30 questions, count pairwise ranking
        for task_idx, (task_name, ratings) in enumerate(sorted_data):
            if ratings['A']:
                model_a = batch_json[task_idx]['Model A']
                rater_organized_data[model_a] += 1
                rater_task_answer[task_name] = 'A'

            elif ratings['B']:
                model_b = batch_json[task_idx]['Model B']
                rater_organized_data[model_b] += 1
                rater_task_answer[task_name] = 'B'

        instruction_dict[i] = rater_organized_data

        first_order = 0
        last_order = 0

        # For first-order or last-order bias, 
        for p in range(15):
            all_pairs_idx = [instruct_idx for instruct_idx in range(len(batch_json)) if batch_json[instruct_idx]['pair_idx'] == p]
            first_pair_idx = all_pairs_idx[0]
            last_pair_idx = all_pairs_idx[1]
            print(f"first: {first_pair_idx+1}, last: {last_pair_idx+1}")
            if (rater_task_answer['task_{}'.format(first_pair_idx+1)] == 'A') and (rater_task_answer['task_{}'.format(last_pair_idx+1)] == 'A'):
                first_order += 1
            else:
                if (rater_task_answer['task_{}'.format(first_pair_idx+1)] == 'B') and (rater_task_answer['task_{}'.format(last_pair_idx+1)] == 'B'):
                    last_order += 1
        
        agg_bias['first-order'].append(round(first_order/15, 3))
        agg_bias['last_order'].append(round(last_order/15, 3))
    
    all_instructions_lst.append(instruction_dict)

    return (all_instructions_lst, agg_bias)
```

```python
def compute_rbo_each(i):

    def sort_dict_values(A):
        sorted_lists = []
        for outer_dict in A:
            for key in outer_dict:
                # Sorting dictionary by values in descending order
                # If values are equal, it retains the order from the original dictionary
                sorted_dict = sorted(outer_dict[key].items(), key=lambda x: x[1], reverse=True)
                sorted_lists.append([item[0] for item in sorted_dict])
        return sorted_lists
    
    sorted_aggregated_raters = sort_dict_values(aggregate_annotators_across_order(i)[0])

    avg_rbo_lst = []
    rbo_matrix = np.zeros((3, 3))
    for a in range(3):
        for b in range(a+1, 3):
            
            rater_a = sorted_aggregated_raters[a]
            rater_b = sorted_aggregated_raters[b]
            all_rbo_scores = rbo.RankingSimilarity(rater_a,rater_b).rbo(p=0.8)
            avg_rbo_lst.append(all_rbo_scores)

    return np.mean(avg_rbo_lst)


def compute_new_iaa_order():
    all_lst = []
    for jj in range(1, 26):
        avg_rbo = compute_rbo_each(jj)
        all_lst.append(avg_rbo)

    return np.mean(all_lst)
```

```python
def calculate_iaa_order(batch_i):
    # for batch_i in range(1, 26):
        
        # Total number of models in each pair
        all_model_sums = dict.fromkeys(model_names, 0)
        with open('human_bias/order/real/random_sampled_examples_order_{}.json'.format(batch_i)) as json_f: 
            original_data = json.load(json_f)
        for example in original_data:
            all_model_sums[example['Model A']] += 1
            all_model_sums[example['Model B']] += 1

        print(all_model_sums)
        
        # Get the counts of each model by annotators 
        agg_instruction, _ = aggregate_annotators_across_order(batch_i)
        agg_instruction_ = agg_instruction[0]

        # Compute selection ratio of model by annotators
        all_sorted_rankings = []
        for i in range(len(agg_instruction_)):
            rater_data = agg_instruction_[i]
            selection_ratio_model = {k: (rater_data[k]/all_model_sums[k]) if all_model_sums[k] != 0 else None for k, _ in rater_data.items()}
            print(selection_ratio_model)
            selection_ratio_model = {k: ratio for k, ratio in selection_ratio_model.items() if ratio is not None}
            print(selection_ratio_model)            
            sorted_model = sorted(selection_ratio_model.keys(), key=lambda key:selection_ratio_model[key], reverse=True)
            print(sorted_model)
            all_sorted_rankings.append(sorted_model)

        # Compute average IAA using RBO
        print(all_sorted_rankings)
        all_rbo_score = []
        for i in range(3):
            for j in range(i+1, 3):
                annotator_i = all_sorted_rankings[i]
                annotator_j = all_sorted_rankings[j]

                rbo_score = rbo.RankingSimilarity(annotator_i, annotator_j).rbo(p=0.8)
                print('rbo score: ', rbo_score)
                if rbo_score != 1:
                    all_rbo_score.append(rbo_score)
        
        avg_rbo = round(sum(all_rbo_score)/len(all_rbo_score), 3)
        return avg_rbo
        
```

```python
def aggregate_annotators_across_salience(batch_i):

    all_instructions_lst = []
    agg_bias = {'shorter': [], 'longer': []}

    # For each instruction, 
    # for batch_i in range(1, 26): 
    path_var_name = f'order_path_{batch_i}'  # Format the variable name
    path_value = eval(path_var_name)   # Get the value of the variable
    valid_raters_data, valid_raters_id = return_valid_data(path_value)

    with open('human_bias/order/real/random_sampled_examples_order_{}.json'.format(batch_i), 'r') as j:
        batch_json = json.load(j)

    instruction_dict = {}
    # For each rater (among 3),
    for i, rater in enumerate(valid_raters_data):
        rater_data = json.loads(rater)[0]
        sorted_data = sorted(rater_data.items(), key=lambda x: int(x[0].split("_")[1]))

        rater_task_answer = {}
        # Create a dictionary to store pairwise counts for each model
        rater_organized_data = dict.fromkeys(model_names, 0)

        # For each of 30 questions, count pairwise ranking
        for task_idx, (task_name, ratings) in enumerate(sorted_data):
            if ratings['A']:
                model_a = batch_json[task_idx]['Model A']
                rater_organized_data[model_a] += 1
                rater_task_answer[task_name] = 'A'

            elif ratings['B']:
                model_b = batch_json[task_idx]['Model B']
                rater_organized_data[model_b] += 1
                rater_task_answer[task_name] = 'B'

        instruction_dict[i] = rater_organized_data

        shorter = 0
        longer = 0
        no_order_bias_pair = 0

        # For salience, 
        for p in range(15):
            all_pairs_idx = [instruct_idx for instruct_idx in range(len(batch_json)) if batch_json[instruct_idx]['pair_idx'] == p]
            first_pair_idx = all_pairs_idx[0]
            last_pair_idx = all_pairs_idx[1]
            if (rater_task_answer['task_{}'.format(first_pair_idx+1)] == 'A') and (rater_task_answer['task_{}'.format(last_pair_idx+1)] ==  'B'):
                no_order_bias_pair += 1
                print('pair id: ', p)
                print('no_order_bias_pairs: ', no_order_bias_pair)
                print(f"first: {first_pair_idx+1}, last: {last_pair_idx+1}")
                print('Choice: A')

                # Calculate 
                response_a = batch_json[first_pair_idx]['System A']
                response_b = batch_json[first_pair_idx]['System B']
                response_a = response_a.replace('\n', '')
                response_b = response_b.replace('\n', '')

                print('response a: ', response_a)
                print('response b: ', response_b)
                # last_response_a = batch_json[last_pair_idx]['System A']
                # last_response_b = batch_json[last_pair_idx]['System B']

                num_tokens_response_a = len(word_tokenize(response_a))
                num_tokens_response_b = len(word_tokenize(response_b))
                print('token num a: ', num_tokens_response_a)
                print('token num b: ', num_tokens_response_b)
                # num_tokens_last_response_a = len(word_tokenize(last_response_a))
                # num_tokens_last_response_b = len(word_tokenize(last_response_b))

                if num_tokens_response_a < num_tokens_response_b:
                    shorter += 1
                
                if num_tokens_response_a > num_tokens_response_b:
                    longer += 1


            else:
                if (rater_task_answer['task_{}'.format(first_pair_idx+1)] == 'B') and (rater_task_answer['task_{}'.format(last_pair_idx+1)] == 'A'):
                    no_order_bias_pair += 1
                    print('pair id: ', p)
                    print('no_order_bias_pairs: ', no_order_bias_pair)
                    print(f"first: {first_pair_idx+1}, last: {last_pair_idx+1}")
                    print('Choice: B')

                    # Calculate
                    response_a = batch_json[first_pair_idx]['System A']
                    response_b = batch_json[first_pair_idx]['System B']
                    response_a = response_a.replace('\n', '')
                    response_b = response_b.replace('\n', '')
                    print('response a: ', response_a)
                    print('response b: ', response_b)
                    num_tokens_response_a = len(word_tokenize(response_a))
                    num_tokens_response_b = len(word_tokenize(response_b))
                    print('token num a: ', num_tokens_response_a)
                    print('token num b: ', num_tokens_response_b)

                    if num_tokens_response_b < num_tokens_response_a:
                        shorter += 1
                    
                    if num_tokens_response_b > num_tokens_response_a:
                        longer += 1
            print('-------------------------------')

        agg_bias['shorter'].append(round(shorter/no_order_bias_pair, 3))
        agg_bias['longer'].append(round(longer/no_order_bias_pair, 3))
    
    all_instructions_lst.append(instruction_dict)

    return (all_instructions_lst, agg_bias)
```
